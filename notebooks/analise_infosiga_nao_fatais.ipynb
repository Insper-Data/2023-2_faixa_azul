```python
## importação das bibliotecas
import pandas as pd
import numpy as np
import pathlib
import os
```

```python
## Caminho do arquivo
nome_arquivo = 'acidentes_naofatais.csv'
path_dir = DATA_DIR = pathlib.Path.cwd().parent / 'dados' / nome_arquivo
print(path_dir) ## print para conferir se o caminho corresponde a pasta e arquivo desejado
```

```python
df = pd.read_csv(path_dir, sep=';', encoding='latin-1') ## leitura do arquivo csv
df.head().T ## visualização do dataframe transposto
```

```python

colunas_relevantes = [
    'Data do Acidente', 'Hora do Acidente', 'Município', 'Logradouro',
    'Numero/KM', 'LAT_(GEO)', 'LONG_(GEO)', 'Veículos Envolvidos - Bicicleta',
       'Veículos Envolvidos - Caminhão', 'Veículos Envolvidos - Automóvel',
       'Veículos Envolvidos - Motocicleta', 'Veículos Envolvidos - Ônibus',
       'Veículos Envolvidos - Pedestre', 'Pessoas Envolvidas - Grave',
       'Pessoas Envolvidas - Ileso', 'Pessoas Envolvidas - Leve',
    'Tipo de Via','Tipo de Acidente - Atropelamento (Pedestre)',
       'Tipo de Acidente - Atropelamento (Animal)',
       'Tipo de Acidente - Choque', 'Tipo de Acidente - Colisão',
       'Tipo de Acidente - Outros tipos de Acidente'
]

df = df[colunas_relevantes]
```

```python
df.head().T
```

```python
df = df[(df['Tipo de Via'] == 'Vias Municipais') & (df['Município'] == 'SAO PAULO')]
```

```python
df.drop(columns=['Tipo de Via', 'Município'], inplace=True)
```

```python

df['LAT_(GEO)'] = pd.to_numeric(df['LAT_(GEO)'].str.replace(',', '.'))
df['LONG_(GEO)'] = pd.to_numeric(df['LONG_(GEO)'].str.replace(',', '.'))
```
