```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('../dados/lentidao.csv', sep=',', encoding='latin-1')
```

```python
df.info()
```

```python
## criar a coluna hora (com um split(' '))
df['hora'] = df['data'].apply(lambda x: x.split(' ')[1] if len(x.split(' ')) >= 2 else '')
```

```python
df.head()
```

```python
df = df.loc[df['hora'] == "12:00"]
```

```python
df.head()
```

```python
df = df.query("corredor == 'Amaro, Av Sto DEC SA'")
```
